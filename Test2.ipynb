```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
office_file.groupby('season').ratings.mean()
```

```python
season_ratings=pd.DataFrame(office_file.groupby('season').ratings.mean())
season_ratings.mean()
```

```python
season_ratings.reset_index(inplace=True)
season_ratings
```

```python
x = season_ratings.season
y = season_ratings.ratings
```

```python
ratingsPerSeason= plt.plot(x,y,'ms--')
plt.xlabel('Season number')
plt.ylabel('Ratings')
plt.title('Ratings per season')
plt.axis([0,10,7,9])
plt.show(ratingsPerSeason)
```

```python
viewsPerSeason= pd.DataFrame(office_file.groupby('season').viewership_mil.mean())
viewsPerSeason.reset_index(inplace=True)
plt.plot(viewsPerSeason.season,viewsPerSeason.viewership_mil,'ms--')
plt.xlabel('Season number')
plt.ylabel('Views in millions')
plt.title('Views per season')
plt.show()
```
